 правило Рунге — разность значений интегралов, рассчитанных по n и 2n точкам, даёт оценку погрешности: $\Delta_{2n} \approx |I_{2n} - I_n|$

```python
def trapezoid_rule(func, interval, accurancy = 0.05, n):
    """Правило трапеций
       rtol - желаемая относительная точность вычислений
       nseg0 - начальное число отрезков разбиения"""
    a = interval[0]
    b = interval[1]
    old_sum = 0
    h = (b - a) / n
    sum_ = (func(a) + func(b))
    for i in range(1, n):
        sum_ += 2*func(a + i * h)
    sum_ *= (h/2)
    
    
    error = max(1, abs(sum_))
    while (error > abs(accurancy * sum_)):
        old_sum = sum_
        sum_ = 0.5 * (sum_ + midpoint_rectangle_rule(func, a, b, n)) # новые точки для уточнения интеграла                                                            # добавляются ровно в середины предыдущих отрезков
        n *= 2
        error = abs(sum_ - old_sum)

    return sum_
```

```python
def my_trap(str_, inter, num):
    a = inter[0]
    b = inter[1]
    h = (b-a)/num
    try:
        sum_ = take_my_formula(str_, a) + take_my_formula(str_, b)
    except ZeroDivisionError:
            a = a + h
            sum_ = take_my_formula(str_, a) + take_my_formula(str_, b)
            
    for i in range(1, num):
        try:
            sum_+=2*take_my_formula(str_, a + i*h)
        except ZeroDivisionError:
            pass
        
    sum_ = sum_*(h/2)
    return sum_
```
